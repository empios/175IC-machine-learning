```
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn import metrics
from mlxtend.plotting import plot_decision_regions
import matplotlib.pyplot as plt

#ZADANIE 1 -> Przygotowanie zestawu danych
iris = datasets.load_iris()
X = iris.data[:, :2]
y = iris.target

#ZADANIE 2 -> podzielenie wartości na trenowane i testowe
x_train,x_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=123)

#ZADANIE 3 -> Utworzenie obiektu SVC
clf = SVC(C=0.5, kernel='linear')
clf.fit(x_train, y_train)
plot_decision_regions(x_train, y_train, clf=clf, legend=2)
y_pred = clf.predict(x_test)
plt.xlabel('Długość [cm]')
plt.ylabel('Szerokość [cm]')
plt.title('C=0.5 oraz kernel=linear')
plt.show()
print("Precyzja:",metrics.accuracy_score(y_test, y_pred))

#ZADANIE 4 -> Kilka przykładów wykorzystania innych kerneli oraz innych wartości C
clf2 = SVC(C=0.01, kernel='poly')
clf2.fit(x_train, y_train)
plot_decision_regions(x_train, y_train, clf=clf2, legend=2)
y_pred2 = clf2.predict(x_test)
plt.xlabel('Długość [cm]')
plt.ylabel('Szerokość [cm]')
plt.title('C=0.01 oraz kernel=poly')
plt.show()
print("Precyzja:",metrics.accuracy_score(y_test, y_pred2))

clf3 = SVC(C=0.244, kernel='sigmoid')
clf3.fit(x_train, y_train)
plot_decision_regions(x_train, y_train, clf=clf3, legend=2)
y_pred2 = clf3.predict(x_test)
plt.xlabel('Długość [cm]')
plt.ylabel('Szerokość [cm]')
plt.title('C=0.244 oraz kernel=sigmoid')
plt.show()
print("Precyzja:",metrics.accuracy_score(y_test, y_pred2))

clf4 = SVC(C=0.80, kernel='rbf')
clf4.fit(x_train, y_train)
plot_decision_regions(x_train, y_train, clf=clf4, legend=2)
y_pred2 = clf4.predict(x_test)
plt.xlabel('Długość [cm]')
plt.ylabel('Szerokość [cm]')
plt.title('C=0.80 oraz kernel=rbf')
plt.show()
print("Precyzja:",metrics.accuracy_score(y_test, y_pred2))



```
